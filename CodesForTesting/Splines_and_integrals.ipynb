```python
from scipy import interpolate
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [1, 3, 4, 3, 5, 7, 5, 6, 8, 7, 8, 9, 8, 7]
n = len(y)
x =range(0,n)
```

```python
plt.plot(x, y, 'ro')
plt.plot(x, y, 'b')
plt.title('Linear interpolation')
plt.show()
```

```python
s = interpolate.InterpolatedUnivariateSpline(x, y)
xfit = np.arange(0, n-1, np.pi/50)
yfit = s(xfit)

plt.plot(x,y,'ro')
plt.plot(xfit, yfit, 'green')
plt.show('Interpolated Univarite Spline')
plt.show()
```

# Computation of integrals

```python
def trapezoidal(f, a, b, n):
  h = float((b-a)/n)
  result = f(a) + f(b)
  for i in range(1,n):
    result += 2*f(a+i*h)
  result *=h/2
  return result
```

```python
from math import exp
v = lambda t: 3*(t**2)*(exp(t**3))
n = 4
```

```python
numerical = trapezoidal(v, 0, 1, n)
numerical
```

```python
V = lambda t: exp(t**3)
exact = V(1)-V(0)
```

```python
numerical - exact
```

```python
numerical = trapezoidal(v, 0, 1, n=100)
```

```python
numerical - exact
```

### Simpsons rule

```python
def simpsons(f, a, b, n):
  h = float(b-a)/n
  result = (f(a)+f(b))
  for i in range(1,n):
    if i%2 ==0:
      result += 4*f(a +i*h)
    else:
      result += 2*f(a +i*h)
  result *=h/3
  return result
```

```python
numerical = simpsons(v, 0, 1, n=1000)
numerical
```

```python
numerical - exact
```
